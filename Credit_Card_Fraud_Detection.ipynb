Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Loading the dataset to a Pandas DataFrame

```python
# Loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv("/content/creditcard.csv")
```

```python
# First 5 rows of Data set
credit_card_data.head()
```

```python
# Dataset information
credit_card_data.info()
```

```python
# Checking in the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# Remove Null values from the data set
credit_card_data = credit_card_data.dropna()
credit_card_data.isnull().sum()
```

```python
# changing the data type of "Class" float into int
credit_card_data['Class'] = credit_card_data['Class'].astype(int)
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

0 --> Normal transation

1 --> Fraudulelment transtion



---



```python
# Separeting for the data for analysis:
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the value of both transactions
credit_card_data.groupby('Class').mean()
```

Under-Sampling

Build a sample dataset containing similler distribution
normal transactions and fraudluent transactions -->  92

```python
legit_sample = legit.sample(n=492)
```

Concatenating two DataFrames

```python
new_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df.groupby('Class').mean()
```

Spleting the data into Features & Targets

```python
X = new_df.drop(columns="Class", axis=1)
Y = new_df['Class']
```

```python
print(X)
```

```python
print(Y)
```

# Split the data into traning data & testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.20, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model traning

Logistic Regression()

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Traning Data
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on Traning data
X_train_prediction = model.predict(X_train)
traning_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on Train data : ", traning_data_accuracy)
```

```python
# Accuracy on test data
X_test_Prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_Prediction,Y_test)
```

```python
print("Accuracy on Test data : " , test_data_accuracy)
```
